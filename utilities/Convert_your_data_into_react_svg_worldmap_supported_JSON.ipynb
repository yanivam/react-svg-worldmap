#### Utility functions

```
apeda_to_world_map_names_mapping = {}  # Your country name to short name mapping.
progress_index = 0  # For continuing mapping where it stops...
```

```
# apeda_to_world_map_names_mapping: APEDA countries name to world map short names mapping
# COUNTRIES_SHORT_NAME_MAPPING: World map countries mapping

def add_new_apeda_country_name_to_mapping(new_country_name: str) -> None:
  """Adds new country name to your mapping.

  Explanation:
    Whenever it finds a new country name, it will try to find the similar country
    name in library's mapping. If a similar one is found, it will add it directly
    into your mapping.
    If it is not sure, it will show the top 10 similar country name for you to
    choose from.
    If it does not find similar ones, it will ask you to enter the short name.
  
  Args:
    new_country_name: The new country name it founded which is not in the mapping.
  """
  similarity_dict = sorted({s: similarity(new_country_name, s) for s in COUNTRIES_SHORT_NAME_MAPPING.keys()}.items(), key=lambda item: item[1], reverse=True)

  if similarity_dict[0][1] > 0.9:
    _short_name = COUNTRIES_SHORT_NAME_MAPPING[similarity_dict[0][0]]
  elif similarity_dict[0][1] > 0.5:
    print(f'\nTop 10 similar ones: {[(c[0], COUNTRIES_SHORT_NAME_MAPPING[c[0]], c[1]) for c in similarity_dict[:10]]}')
    _short_name = str(input(f'short name for {new_country_name}:'))
  else:
    _short_name = str(input(f'short name for {new_country_name}:'))

  apeda_to_world_map_names_mapping[new_country_name] = _short_name


def convert_countries_string_to_short_names(string: str):
  """Convert your country data into react-svg-worldmap supported mapping
  
  Args:
    string: A long string including country name & values. i.e.
    1	U S A	853,090.00	659,119.00
    2	Germany	261,564.00	454,509.00
    ...
  
  Output:
    It will print JSON equivalent result that you can put directly in a new JSON
    file.
  """
  str_l = string.split('\n')
  results = []
  included_countries = []
  results.append('{\n\t"data": [')

  for s in str_l:
    sl = s.split()
    name = ' '.join(sl[1:-2])

    # If name is not present, add it.
    if name not in apeda_to_world_map_names_mapping:
      add_new_apeda_country_name_to_mapping(name)

    short_name = apeda_to_world_map_names_mapping[name]
    included_countries.append(short_name)
    val = sl[-2].replace(',', '')
    results.append('\t\t{ ' + f'"country": "{short_name}", "value": {val}' + ' },')

  remaining_countries = list(set(COUNTRIES_SHORT_NAME_MAPPING.values()) - set(included_countries))
  for s in remaining_countries:
    results.append('\t\t{ ' + f'"country": "{s}", "value": 0' + ' },')
  
  results[-1] = results[-1][:-1]
  results.append('\t]\n}')

  for r in results:
    print(r)
```
